```python
import numpy as np
from skimage import io

image = np.array([np.load(f'images/car_{x}.npy') for x in range(9)])
print(image.shape)
```

```python
suma_totala = 0
for i in range(9):
    suma_curenta = np.sum(image[i])
    print(f'Imaginea {i+1} are suma {suma_curenta}')
    suma_totala += suma_curenta
print(f'Suma totala este {suma_totala}')
```

```python
images_sums = [np.sum(image[i]) for i in range(9)]
print(*images_sums)

print(f'Suma totala este {np.sum(images_sums)}')

poz_max = np.argmax(images_sums)
print(f'Imaginea cu suma maxima este {poz_max+1}')
```

```python
imagine_medie = np.mean(image, axis=0)
io.imshow(imagine_medie.astype(np.uint8)) # petru a putea fi afisata
io.show()
```

```python
deviatie = np.std(image)
print(deviatie)
```

```python
images_normalized = [(imag - imagine_medie) / deviatie for imag in image]
io.imshow(images_normalized[0].astype(np.uint8))
io.show()
```

```python
# Decupai fiecare imagine, afind numai liniile cuprinse ntre 200 i 300, respectiv coloanele cuprinse ntre 280 i 400.
imagini_decupate = [imag[200:300, 280:400] for imag in image]
print(np.shape(imagini_decupate))

io.imshow(imagini_decupate[0].astype(np.uint8))
io.show()
```
